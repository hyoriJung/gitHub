### 2. Exercise 63:Temperature Conversion Table

```python
def convertF(num): #화씨
    ftemp=(num-32.0)*5.0/9.0
    return ftemp 

def convertC(num): #섭씨
    ctemp=(num*9.0/5.0)+32
    return ctemp 
```

```python
for i in range(0,101):
    a=convertF(i)
    b=convertC(i)
    print('입력값 : ',i,',화씨 : ',a,',섭씨 : ',b)
```

### 3. Exercise 66: Compute a Grade Point Average

```python
sum=0
count=0
st= {"A+": 4.0, "A": 4.0, "A-": 3.7, "B+": 3.3, "B": 3.0, "B-": 2.7,
    "C+": 2.3, "C": 2.0, "C-": 1.7,"D+": 1.3, "D": 1.0,"F": 0.0}
    
while True:
    grade=input("학점을 입력하십시오.(빈줄일 경우 종료) : ")
    if grade is '':
        break
    else:
        sum+=st[grade]
        count+=1
        
print(sum/count)        
```

### 4. Exercise 67: Admission Price

```python
def convertAge(age):
    cost=0;
    if age>=65:
        cost=18.00
    elif 65>age>=13:
        cost=23.00
    elif 13>age>=3:
        cost=14.00
    else:
        cost=0.00 
    return cost    

cost=0;

while True:
    age=input("당신의 나이를 입력하십시오.(빈줄일 경우 종료) : ")
    if age is '':
        break
    else:
        c=convertAge(float(age))
        cost+=c
        
print("총 입장료는 : ",format(cost,"10.2f"))        
        
```

### 5. Exercise 72: Is a String a Palindrome?

```python
def isPalindrome(word):
    list_word = list(word)
    print(list_word)
    for i in range(0, len(list_word) // 2):

        if list_word[i] == list_word[len(list_word) - 1 - i]:
            continue
        else:
            return False

    return True

word=input("단어를 입력하세요: ")
isPalindrome(word)
```

### 7. Exercise 75: Greatest Common Divisor

```python
a=int(input("첫번째 양수를 입력하세요 : "))
b=int(input("두번째 양수를 입력하세요 : "))

def gcmMethod(a,b):
    gcm=1
    
    for k in range(2,min(a,b)+1):
        while(a%k==0) & (b%k==0):
            a=a//k
            b=b//k
            gcm=gcm*k
            continue
    return gcm

gcmMethod(a,b)
```

### 8. Exercise 78: Decimal to Binary

```python
def dec2bin(q):
    r = ""
    if q == 0 :
        return '0'

    while q > 0 :
        r = str( q % 2 ) + r
        q = q >> 1

    return r

n = int(input("2진수로 변환할 10진수를 입력하세요 : "))
dec2bin(n)
```
